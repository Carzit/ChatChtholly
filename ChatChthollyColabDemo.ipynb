### Instruction
The following cell is the main program of ChatChtholly. You can experience a conversation with Chtholly.

- Click the Run button, run the cell, and start initialization. “Chtholly is ready! Now you can chat with her.” When the word appears, it finished initialization and you will enter the conversation mode with Chtholly.
When "You:" appears, you can enter the text you want to send to Chtholly on the next line. Press Enter to send.

- When the words "Text Generating..." appear, text generation is in progress.
Once generated, you'll see Chtholly's reply under the word "Chtholly:", rendered in the original language and its Japanese translation.

- When the words "Voice Generating..." appear, speech generation is in progress.
When the build is complete, "Successfully saved!" appears. Words. An audio playback window will then be generated, and you can hear the corresponding voice of the reply by clicking the play button (in the voice of Chtholly!). ）

- As above, if you want to talk again, you can run the cell again. (Chtholly can remember what you once said!) )
